## **DEMANDA Y TEMPERATURA:** AÑOS 2020-2022

#### **1) Importamos libreria útiles:**


```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime

sns.set(style="darkgrid")
maxScatter = 3000
```

#### **2) Cargamos datos:**

- Comprobamos que trabajamos en el directorio correcto y cargamos datos de carpeta "data":

```python
dir = os.getcwd()
DIR = dir
print('Current working directory:', os.getcwd())


# Nos guardamos los datos que queremos en un dataframe usando pandas
dataframe = pd.read_csv('..\data\G02_DATOS_DEM_G1.csv', sep=',', verbose=True)
dataframe.FECHA = pd.to_datetime(dataframe.FECHA, format='%d/%m/%Y')
dataframe
```

- En este analisis vamos a centrarnos en analizar los años 2020, 2021, 2022. Por tanto creamos un dataframe que únicamente contenga los datos de estos años.

```python
dataframe_202122 = dataframe[dataframe['ANNO'] >= 2020]
dataframe_202122
```

##### **4.2) Evolución temporal (años 2020-2022):**

```python
fig, (ax1, ax2) = plt.subplots(2, figsize=(15, 8)) 

ax1.plot(dataframe['FECHA'], dataframe['DEM'], label = 'DEM (2014-2022)', color = 'seagreen')
ax1.axvspan(datetime.date(2020,1,1), datetime.date(2022,2,2), label = 'DEM (2020-2022)', color='seagreen', alpha=0.3)
ax1.set_xlim([datetime.date(2014, 1, 1), datetime.date(2022, 2, 2)])

ax2.plot(dataframe_202122['FECHA'], dataframe_202122['DEM'], label = 'DEM (2020-2022)', color = 'seagreen')
ax2.set_xlim([datetime.date(2020, 1, 1), datetime.date(2022, 2, 2)])

fig.suptitle("Evolución temporal de la DEMANDA (2020-2022):", fontsize = 20)
ax1.set_ylabel("demanda (GWh)", fontsize = 10)
ax2.set_ylabel("demanda (GWh)", fontsize = 10)
ax1.legend()
ax2.legend()
plt.grid()
plt.savefig('../reports/figures/Evolución temporal de la DEMANDA (2020-2022).png', bbox_inches='tight')
plt.show()
```

- ¿Por qué seleccionamos estos años? Comparémoslos con los anteriores:

        Para compararlo podemos coger la media de los anteriores años...        

```python
dataframe_ans = dataframe[dataframe['ANNO'] < 2020]
dataframe_ans
```

```python
dataframe_ans.describe(include=float)
```

```python
media_ans = dataframe_ans['DEM'].mean()
print('Demanda media años anteriores:', media_ans)
```

        Y tomarla como el 0 de nuestra nueva gráfica para ver las subidas y bajadas de demanda más fácilmente:

```python
def comparar_con_media_ans(x):
    return x - media_ans

cdataframe_202122 = dataframe_202122.copy()
cdataframe_202122['DEM'] = dataframe_202122['DEM'].apply(comparar_con_media_ans)
cdataframe_202122
```

        Cogemos los 3 años anteriores de referencia para comparar (comparándolo igualmente con la misma media):

```python
mask = (dataframe['ANNO'] > 2016) & (dataframe['ANNO'] < 2020)
dataframe_171819 = dataframe.loc[mask]

# Comparamos con la media
cdataframe_171819 = dataframe.copy()
cdataframe_171819['DEM'] = dataframe_171819['DEM'].apply(comparar_con_media_ans)
cdataframe_171819
```

```python
sns.axes_style("darkgrid")
fig, (ax1, ax2) = plt.subplots(2, figsize=(15, 8))
# Ajustamos el espaciado
fig.subplots_adjust(hspace = 0.4)

# Demanda 2017 - 2019
ax1.fill_between(cdataframe_171819['FECHA'], cdataframe_171819['DEM'], where=cdataframe_171819['DEM']<0,
                color = 'red', label = 'DEM (2017 - 2019) - negativa', alpha = 0.3)

ax1.fill_between(cdataframe_171819['FECHA'], cdataframe_171819['DEM'], where=cdataframe_171819['DEM']>0,
                color = 'seagreen', label = 'DEM (2017 - 2019) - positiva', alpha = 0.3)

# Demanda 2020 - 2022
ax2.fill_between(cdataframe_202122['FECHA'], cdataframe_202122['DEM'], where=cdataframe_202122['DEM']<0,
                color = 'red', label = 'DEM (2020 - 2022) - negativa', alpha = 0.3)

ax2.fill_between(cdataframe_202122['FECHA'], cdataframe_202122['DEM'], where=cdataframe_202122['DEM']>0,
                color = 'seagreen', label = 'DEM (2020 - 2022) - positiva', alpha = 0.3)

ax1.set_title("Evolución temporal de la DEMANDA desde 2017 hasta 2019:", fontsize = 20)
ax2.set_title("Evolución temporal de la DEMANDA desde 2020 hasta 2022:", fontsize = 20)

# Sólo seleccionamos hasta el 2 de febrero, ya que no tenemos más datos de 2022 y queremos
# que haya tantos datos en la gráfica de arriba como en la de abajo (para comparar fácilmente)
ax1.set_xlim([datetime.date(2017, 1, 1), datetime.date(2019, 2, 2)])
ax2.set_xlim([datetime.date(2020, 1, 1), datetime.date(2022, 2, 2)])

ax1.set_ylabel("demanda (GWh)", fontsize = 10)
ax2.set_ylabel("demanda (GWh)", fontsize = 10)

ax1.legend()
ax2.legend()

plt.savefig('../reports/figures/Evolución temporal de la DEMANDA (comparativa).png', bbox_inches='tight')
plt.show()
```

- Comparémos las gráficas de 2017-2019 y de 2020-2022 una encima de la otra:

```python
# Primero creamos un dataframe que esté desfasado 3 años:
from dateutil.relativedelta import relativedelta

def add_3_year(x):
    return x + relativedelta(years=3)

def add_three(x):
    return x + 1

bdataframe_171819 = cdataframe_171819.copy()
bdataframe_171819['FECHA'] = bdataframe_171819['FECHA'].apply(add_3_year)
bdataframe_171819['ANNO'] = bdataframe_171819['ANNO'].apply(add_three)

# No hace falta acotar el dataframe de 2017-19 (con el desfase de 2020-22),
# ya que lo podremos hacer luego al graficar.
bdataframe_171819
```

```python
sns.axes_style("darkgrid")
fig = plt.figure(2, figsize=(15, 4))

# Demanda 2020 - 2022
plt.fill_between(cdataframe_202122['FECHA'], cdataframe_202122['DEM'], where=cdataframe_202122['DEM']<0,
                color = 'red', label = 'DEM (2020 - 2022) - negativa', alpha = 0.5)

plt.fill_between(cdataframe_202122['FECHA'], cdataframe_202122['DEM'], where=cdataframe_202122['DEM']>0,
                color = 'seagreen', label = 'DEM (2020 - 2022) - positiva', alpha = 0.5)

# Demanda 2017 - 2019
plt.fill_between(bdataframe_171819['FECHA'], bdataframe_171819['DEM'], where=bdataframe_171819['DEM'],
                color = 'gray', label = 'DEM (2017 - 2019) - desfasado')

plt.title("Evolución temporal de la DEMANDA (comparación años):", fontsize = 20)
plt.xlim([datetime.date(2020, 1, 1), datetime.date(2022, 2, 2)])
plt.ylabel("demanda (GWh)", fontsize = 10)
plt.legend()

plt.savefig('../reports/figures/Evolución temporal de la DEMANDA (comparativa 2).png', bbox_inches='tight')
plt.show()
```
